### Developing a Scoring Function for Best Posting Hours

Now, let's create a simple scoring function to identify and rank the best posting hours based on the engagement metrics. This function will normalize the metrics and combine them into a single score, allowing us to see which hours generally yield the highest engagement.

````python
from sklearn.preprocessing import MinMaxScaler

def rank_posting_hours(df, engagement_metrics):
    """
    Ranks posting hours based on a combined engagement score.

    Args:
        df (pd.DataFrame): The input DataFrame containing 'hour_posted' and engagement metrics.
        engagement_metrics (list): A list of column names representing engagement metrics.

    Returns:
        pd.DataFrame: A DataFrame with 'hour_posted', combined 'engagement_score', and 'rank'.
    """

    # Calculate hourly average for each engagement metric
    hourly_avg_metrics = df.groupby('hour_posted')[engagement_metrics].mean()

    # Normalize each engagement metric
    scaler = MinMaxScaler()
    normalized_metrics = scaler.fit_transform(hourly_avg_metrics)
    normalized_df = pd.DataFrame(normalized_metrics, columns=engagement_metrics, index=hourly_avg_metrics.index)

    # Calculate a combined engagement score (simple sum of normalized metrics)
    normalized_df['engagement_score'] = normalized_df[engagement_metrics].sum(axis=1)

    # Rank the hours by the engagement score
    ranked_hours = normalized_df.sort_values(by='engagement_score', ascending=False).reset_index()
    ranked_hours['rank'] = ranked_hours.index + 1

    return ranked_hours[['hour_posted', 'engagement_score', 'rank']]

# Example usage of the scoring function
ranked_posting_hours_df = rank_posting_hours(twitter_df, engagement_metrics)

print('\nRanked Posting Hours based on Combined Engagement Score:')
display(ranked_posting_hours_df.head(10))
````

## Preparing Data for XGBoost Engagement Predictor

To build the XGBoost engagement predictor, we first need to prepare our `twitter_df` dataset. This involves several steps:

1.  **Create a Combined Engagement Score**: We'll normalize and sum the individual engagement metrics (`replies`, `reposts`, `likes`, `views`, `quotes`, `bookmarks`) to create a single target variable.
2.  **Log-Transform Engagement Score**: As requested, we will apply a log transformation (`log1p`) to the combined engagement score to handle potential skewness and make the distribution more Gaussian-like, which often improves model performance.
3.  **Feature Engineering**: We'll extract and prepare features from the `twitter_df`. Specifically, we'll convert categorical features like `day_of_week_posted` and `hour_posted` into a numerical format suitable for XGBoost, typically using one-hot encoding.
4.  **Chronological Train/Test Split**: We'll sort the data by `date_posted` and split it chronologically to ensure earlier posts are used for training and later posts for testing, mimicking a real-world prediction scenario.

````python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 1. Create a Combined Engagement Score
# Normalize each engagement metric (excluding 'views' as it tends to be much larger and could dominate)
# I'll re-evaluate if 'views' should be included or scaled differently if the initial model struggles.

# Let's use the 'engagement_metrics' list defined earlier.
# Filter out 'views' from normalization for the combined score if it skews results too much.
# For now, let's normalize all of them and sum, similar to the `rank_posting_hours` function.

# Make a copy to avoid SettingWithCopyWarning
twitter_df_processed = twitter_df.copy()

scaler = MinMaxScaler()
normalized_metrics = scaler.fit_transform(twitter_df_processed[engagement_metrics])
normalized_df = pd.DataFrame(normalized_metrics, columns=[f'{col}_normalized' for col in engagement_metrics], index=twitter_df_processed.index)

twitter_df_processed['combined_engagement_score'] = normalized_df.sum(axis=1)

# 2. Log-Transform Engagement Score
twitter_df_processed['log_combined_engagement_score'] = np.log1p(twitter_df_processed['combined_engagement_score'])

print("DataFrame with combined and log-transformed engagement scores:")
display(twitter_df_processed[['combined_engagement_score', 'log_combined_engagement_score']].head())
````

````python
# 3. Feature Engineering: One-hot encode categorical features

# Ensure 'hour_posted' and 'day_of_week_posted' are treated as categorical for one-hot encoding
twitter_df_processed['hour_posted'] = twitter_df_processed['hour_posted'].astype('category')
twitter_df_processed['day_of_week_posted'] = twitter_df_processed['day_of_week_posted'].astype('category')

# Select features for the model
# Include original numerical features and new categorical ones

# List of initial features for XGBoost. We will iterate on this later.
# For a baseline, let's include basic user info and the time features.
features = ['followers', 'posts_count', 'following', 'is_verified', 'hour_posted', 'day_of_week_posted']

# One-hot encode categorical features
twitter_df_encoded = pd.get_dummies(twitter_df_processed[features], columns=['hour_posted', 'day_of_week_posted'], drop_first=True)

# Display the first few rows of the encoded features
print("Encoded features for XGBoost:")
display(twitter_df_encoded.head())
print(f"Shape of encoded features: {twitter_df_encoded.shape}")
````

````python
# 4. Chronological Train/Test Split

# Sort the DataFrame by 'date_posted' to ensure chronological split
twitter_df_processed = twitter_df_processed.sort_values(by='date_posted').reset_index(drop=True)

X = twitter_df_encoded
y = twitter_df_processed['log_combined_engagement_score']

# Define the split point (e.g., 80% for training, 20% for testing)
split_point = int(len(twitter_df_processed) * 0.8)

X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

print(f"Training set size: {len(X_train)} samples")
print(f"Test set size: {len(X_test)} samples")
print(f"First training date: {twitter_df_processed['date_posted'].iloc[0]}")
print(f"Last training date: {twitter_df_processed['date_posted'].iloc[split_point-1]}")
print(f"First test date: {twitter_df_processed['date_posted'].iloc[split_point]}")
print(f"Last test date: {twitter_df_processed['date_posted'].iloc[-1]}")
````

## Training a Baseline XGBoost Model for Engagement Prediction

Now that our data is prepared, we will train a baseline XGBoost Regressor model to predict the `log_combined_engagement_score`. We'll start with default parameters to establish a baseline performance, which we will later improve through hyperparameter tuning and feature selection.

We will evaluate the model using Root Mean Squared Error (RMSE) on the log-transformed engagement score, as specified. This metric is suitable for regression tasks and will help us understand the average magnitude of the errors in our predictions.

````python
import xgboost as xgb
from sklearn.metrics import mean_squared_error

# Initialize and train the XGBoost Regressor with default parameters
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = xgb_model.predict(X_test)

# Evaluate the model using RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Baseline XGBoost Model RMSE on log-transformed engagement: {rmse:.4f}")

# Display feature importances
print("\nFeature Importances:")
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': xgb_model.feature_importances_
}).sort_values(by='Importance', ascending=False)
display(importance_df.head(10))
````

## Hyperparameter Tuning for XGBoost Model

To further improve the performance of our XGBoost model, we will perform hyperparameter tuning using `GridSearchCV`. This process will systematically search through a predefined range of hyperparameters to find the combination that yields the best performance on our validation set (or cross-validation folds). We'll focus on key parameters such as `n_estimators`, `max_depth`, `learning_rate`, and `subsample`.

The evaluation metric will remain Root Mean Squared Error (RMSE) on the log-transformed engagement score.

````python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300], # Number of boosting rounds
    'max_depth': [3, 5, 7],         # Maximum depth of a tree
    'learning_rate': [0.01, 0.1, 0.2], # Step size shrinkage to prevent overfitting
    'subsample': [0.7, 0.8, 0.9]     # Subsample ratio of the training instance
}

# Initialize the XGBoost Regressor
estimator = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, use_label_encoder=False, eval_metric='rmse')

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    scoring='neg_root_mean_squared_error', # GridSearchCV minimizes, so we use negative RMSE
    cv=3, # 3-fold cross-validation
    verbose=1, # Print progress messages
    n_jobs=-1  # Use all available cores
)

# Perform the grid search on the training data
print("Starting hyperparameter tuning...")
grid_search.fit(X_train, y_train)

print("Hyperparameter tuning complete.")

# Get the best parameters and the best score
best_params = grid_search.best_params_
best_rmse_score = -grid_search.best_score_ # Convert back to positive RMSE

print(f"\nBest Hyperparameters found: {best_params}")
print(f"Best RMSE from GridSearchCV (cross-validation): {best_rmse_score:.4f}")

# Train the model with the best parameters
optimized_xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, use_label_encoder=False, eval_metric='rmse', **best_params)
optimized_xgb_model.fit(X_train, y_train)

# Make predictions on the test set with the optimized model
y_pred_optimized = optimized_xgb_model.predict(X_test)

# Evaluate the optimized model on the test set using RMSE
rmse_optimized = np.sqrt(mean_squared_error(y_test, y_pred_optimized))
print(f"Optimized XGBoost Model RMSE on log-transformed engagement (test set): {rmse_optimized:.4f}")

# Display feature importances for the optimized model
print("\nFeature Importances for Optimized Model:")
importance_df_optimized = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': optimized_xgb_model.feature_importances_
}).sort_values(by='Importance', ascending=False)
display(importance_df_optimized.head(10))
````

````python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300], # Number of boosting rounds
    'max_depth': [3, 5, 7],         # Maximum depth of a tree
    'learning_rate': [0.01, 0.1, 0.2], # Step size shrinkage to prevent overfitting
    'subsample': [0.7, 0.8, 0.9]     # Subsample ratio of the training instance
}

# Initialize the XGBoost Regressor
estimator = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, use_label_encoder=False, eval_metric='rmse')

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    scoring='neg_root_mean_squared_error', # GridSearchCV minimizes, so we use negative RMSE
    cv=3, # 3-fold cross-validation
    verbose=1, # Print progress messages
    n_jobs=-1  # Use all available cores
)

# Perform the grid search on the training data
print("Starting hyperparameter tuning...")
grid_search.fit(X_train, y_train)

print("Hyperparameter tuning complete.")

# Get the best parameters and the best score
best_params = grid_search.best_params_
best_rmse_score = -grid_search.best_score_ # Convert back to positive RMSE

print(f"\nBest Hyperparameters found: {best_params}")
print(f"Best RMSE from GridSearchCV (cross-validation): {best_rmse_score:.4f}")

# Train the model with the best parameters
optimized_xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, use_label_encoder=False, eval_metric='rmse', **best_params)
optimized_xgb_model.fit(X_train, y_train)

# Make predictions on the test set with the optimized model
y_pred_optimized = optimized_xgb_model.predict(X_test)

# Evaluate the optimized model on the test set using RMSE
rmse_optimized = np.sqrt(mean_squared_error(y_test, y_pred_optimized))
print(f"Optimized XGBoost Model RMSE on log-transformed engagement (test set): {rmse_optimized:.4f}")

# Display feature importances for the optimized model
print("\nFeature Importances for Optimized Model:")
importance_df_optimized = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': optimized_xgb_model.feature_importances_
}).sort_values(by='Importance', ascending=False)
display(importance_df_optimized.head(10))
````

## Developing a Recommender Function for Best Posting Hours

Based on the `rank_posting_hours` function and the `ranked_posting_hours_df` DataFrame we previously created, we can now develop a simple recommender function. This function will take a platform as input (though for now, it will use the Twitter-derived insights) and suggest the best posting time(s) based on the combined engagement score.

For future expansion, this function could be extended to handle different platforms if more platform-specific `ranked_posting_hours_df` like DataFrames were available.

````python
def get_best_posting_time(platform="Twitter", top_n=3):
    """
    Recommends the best posting time(s) based on the ranked engagement scores.

    Args:
        platform (str): The social media platform (currently only 'Twitter' is supported).
        top_n (int): The number of top hours to recommend.

    Returns:
        list: A list of recommended hours (integers) or an error message if the platform is not supported.
    """
    if platform == "Twitter":
        if 'ranked_posting_hours_df' in globals():
            top_hours = ranked_posting_hours_df.head(top_n)['hour_posted'].tolist()
            return top_hours
        else:
            return "Error: ranked_posting_hours_df not found. Please run the ranking analysis first."
    else:
        return f"Platform '{platform}' is not currently supported for posting time recommendations."

# Example usage:
print("Top 3 best posting hours for Twitter:")
display(get_best_posting_time(platform="Twitter", top_n=3))

print("Top 1 best posting hour for Twitter:")
display(get_best_posting_time(platform="Twitter", top_n=1))

print("Attempting to get recommendations for an unsupported platform:")
display(get_best_posting_time(platform="LinkedIn"))
````
