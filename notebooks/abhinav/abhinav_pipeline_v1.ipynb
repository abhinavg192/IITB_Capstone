````python
import json

json_template = PromptTemplate(
    input_variables=["brand_name", "topic", "tone", "platform"],
    template="""
You are an expert social media content creator.

Generate exactly 5 distinct social media post variants for the brand {brand_name}.

PLATFORM: {platform}
TOPIC: {topic}
TONE: {tone}

PLATFORM RULES:
- Twitter: maximum 280 characters, 2-3 hashtags, punchy and direct
- LinkedIn: 150-300 words, professional but human, 3-5 hashtags
- Instagram: storytelling style, 5-10 hashtags, strong opening hook

IMPORTANT: You must respond with ONLY a valid JSON array.
No explanations. No preamble. No markdown. Just the JSON.

Return exactly this structure:
[
  {{
    "variant_id": 1,
    "post_text": "the post content here",
    "hashtags": ["#hashtag1", "#hashtag2"],
    "tone": "describe the tone used",
    "suggested_posting_time": "best day and time to post",
    "platform": "{platform}"
  }},
  {{
    "variant_id": 2,
    "post_text": "the post content here",
    "hashtags": ["#hashtag1", "#hashtag2"],
    "tone": "describe the tone used",
    "suggested_posting_time": "best day and time to post",
    "platform": "{platform}"
  }},
  {{
    "variant_id": 3,
    "post_text": "the post content here",
    "hashtags": ["#hashtag1", "#hashtag2"],
    "tone": "describe the tone used",
    "suggested_posting_time": "best day and time to post",
    "platform": "{platform}"
  }},
  {{
    "variant_id": 4,
    "post_text": "the post content here",
    "hashtags": ["#hashtag1", "#hashtag2"],
    "tone": "describe the tone used",
    "suggested_posting_time": "best day and time to post",
    "platform": "{platform}"
  }},
  {{
    "variant_id": 5,
    "post_text": "the post content here",
    "hashtags": ["#hashtag1", "#hashtag2"],
    "tone": "describe the tone used",
    "suggested_posting_time": "best day and time to post",
    "platform": "{platform}"
  }}
]
"""
)

print("✅ JSON template loaded!")
````

````python
# Generate 5 variants as structured JSON
filled_json_prompt = json_template.format(
    brand_name="Adobe",
    topic="Adobe's acquisition of Semrush",
    tone="professional, strategic, reassuring to investors",
    platform="linkedin"
)

# Call Claude
response = llm.invoke([HumanMessage(content=filled_json_prompt)])

# Parse the JSON response
try:
    # Clean the response in case Claude adds any extra text
    response_text = response.content.strip()

    # Remove markdown code blocks if Claude adds them
    if response_text.startswith("```"):
        response_text = response_text.split("```")[1]
        if response_text.startswith("json"):
            response_text = response_text[4:]

    # Parse into Python list
    variants = json.loads(response_text)

    print("✅ JSON parsed successfully!")
    print(f"Number of variants: {len(variants)}")
    print("\n--- VARIANT 1 PREVIEW ---")
    print(f"Post text: {variants[0]['post_text'][:100]}...")
    print(f"Hashtags: {variants[0]['hashtags']}")
    print(f"Tone: {variants[0]['tone']}")
    print(f"Posting time: {variants[0]['suggested_posting_time']}")
    print(f"Platform: {variants[0]['platform']}")

except json.JSONDecodeError as e:
    print(f"❌ JSON parsing failed: {e}")
    print("Raw response:")
    print(response.content)
````

````python
# Print all 5 variants cleanly
for i, variant in enumerate(variants, 1):
    print(f"{'='*50}")
    print(f"VARIANT {i}")
    print(f"{'='*50}")
    print(f"Post Text:\n{variant['post_text']}")
    print(f"\nHashtags: {' '.join(variant['hashtags'])}")
    print(f"Tone: {variant['tone']}")
    print(f"Suggested Posting Time: {variant['suggested_posting_time']}")
    print(f"Platform: {variant['platform']}")
    print()
````

````python
from transformers import pipeline

print("Loading DistilBERT sentiment model...")
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
print("✅ DistilBERT loaded!")
````

````python
import re

def get_sentiment_score(text):
    """
    Returns sentiment score from -1 to 1
    using DistilBERT
    """
    result = sentiment_analyzer(text[:512])[0]
    score = result['score']
    if result['label'] == 'POSITIVE':
        return round(score, 4)
    else:
        return round(-score, 4)

def extract_features(variant, platform):
    """
    Extracts content quality features only.
    Posting time is returned separately
    as a recommendation — not a scoring feature.
    """
    post_text = variant['post_text']
    hashtags = variant['hashtags']

    # Content quality features only
    caption_length = len(post_text)
    hashtag_count = len(hashtags)
    sentiment_score = get_sentiment_score(post_text)

    cta_keywords = [
        'click', 'learn more', 'discover', 'explore',
        'try', 'get', 'join', 'sign up', 'download',
        'shop', 'buy', 'visit', 'check out', 'read',
        'watch', 'follow', 'share', 'comment', 'tag'
    ]
    has_cta = int(any(
        keyword in post_text.lower()
        for keyword in cta_keywords
    ))

    platform_map = {"twitter": 0, "linkedin": 1, "instagram": 2}
    platform_encoded = platform_map.get(platform, 0)

    word_count = len(post_text.split())
    has_question = int('?' in post_text)

    emoji_pattern = re.compile(
        "["u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F9FF"
        u"\U00002700-\U000027BF""]+",
        flags=re.UNICODE
    )
    has_emoji = int(bool(emoji_pattern.search(post_text)))

    return {
        "caption_length": caption_length,
        "hashtag_count": hashtag_count,
        "sentiment_score": sentiment_score,
        "has_cta": has_cta,
        "platform": platform_encoded,
        "word_count": word_count,
        "has_question": has_question,
        "has_emoji": has_emoji
        # posting_hour REMOVED from scoring features
    }
````
