```python
# Load API keys securely from Colab Secrets
from google.colab import userdata

ANTHROPIC_API_KEY = userdata.get('ANTHROPIC_API_KEY')
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

# Quick check - confirms keys loaded without exposing them
print("Anthropic key loaded:", ANTHROPIC_API_KEY[:10] + "...")
print("OpenAI key loaded:", OPENAI_API_KEY[:10] + "...")
```

```python
# First Claude API call using LangChain
from langchain_anthropic import ChatAnthropic
from langchain_core.messages import HumanMessage

# Initialize the Claude model
llm = ChatAnthropic(
    model="claude-haiku-4-5-20251001",
    api_key=ANTHROPIC_API_KEY,
    max_tokens=1000
)

# Send a simple test message
response = llm.invoke([HumanMessage(content="Write a short Twitter post about coffee.")])

print(response.content)
```

```python
from langchain_core.prompts import PromptTemplate

linkedin_template = PromptTemplate(
    input_variables=["brand_name", "topic", "tone"],
    template="""
You are an expert social media content creator.

Your task is to write 5 distinct LinkedIn post variants for the brand {brand_name}.

TOPIC: {topic}
TONE: {tone}

LinkedIn RULES:
- Maximum 600 characters per post
- Use 2-3 relevant hashtags per post
- Keep language punchy and direct
- Include a call to action where natural

OUTPUT FORMAT:
Return exactly 5 posts numbered like this:
1. [post text] [hashtags]
2. [post text] [hashtags]
3. [post text] [hashtags]
4. [post text] [hashtags]
5. [post text] [hashtags]

Write only the posts. No explanations.
"""
)

# Test it by filling in the variables
filled_prompt = linkedin_template.format(
    brand_name="Adobe",
    topic="Acqusition of semrush",
    tone="innovation, bold, empowering, inspiring, genuine, professional, market leading "
)

print(filled_prompt)
```

```python
# Send the filled prompt to Claude and get 5 Twitter variants
response = llm.invoke([HumanMessage(content=filled_prompt)])

print(response.content)
```
